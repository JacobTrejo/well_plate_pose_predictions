```python
import numpy as np
import matplotlib.pyplot as plt


arr = np.load('full_cc_list.npy')
fig, ax = plt.subplots()
ax.violinplot(arr)
plt.title('Correlation Coefficients From Well Plate Data')
ax.set_ylabel('Correlation Coefficients')
ax.xaxis.set_tick_params(labelbottom=False)
plt.ylim(-.2, 1)
plt.show()
```
